```python
df = pd.read_csv("filtered_sarcasm_data.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each label
label_counts = df['label'].value_counts()

# Plot the distribution of sarcastic vs. non-sarcastic comments
plt.figure(figsize=(6, 4))
plt.bar(label_counts.index, label_counts.values, tick_label=["Non-Sarcastic", "Sarcastic"])
plt.xlabel("Label")
plt.ylabel("Count")
plt.title("Distribution of Sarcastic vs. Non-Sarcastic Comments")
plt.show()
```

```python
import seaborn as sns

# Create a boxplot comparing sarcastic vs. non-sarcastic comment lengths
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["label"], y=df["comment_length"])
plt.xticks([0, 1], ["Non-Sarcastic", "Sarcastic"])
plt.xlabel("Label")
plt.ylabel("Comment Length")
plt.title("Comment Length Distribution for Sarcastic vs. Non-Sarcastic Comments")
plt.show()
```
